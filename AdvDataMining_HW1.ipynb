## **(b): Replace missing values with mean values**

````python
# function to replace null values with mean for numeric features
def replace_nan_with_mean(df):
  for col in df.columns:
    if df[col].dtype in ["int64", "float64"]:
      df[col].fillna(df[col].mean(), inplace=True)
  return df

# apply function on mcdonals data
filled_mcdonals_menu_df = replace_nan_with_mean(mcdonals_menu_df)
````

# **Linear Regression**

````python
# import required libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.preprocessing import StandardScaler
````

````python
# select the required features
mcdonals_df_linear_reg = mcdonals_menu_df[["Calories", "Carbohydrates", "Protein", "Total Fat"]]
mcdonals_df_linear_reg.head()
````

## **(a):  Train a predictor to predict the Calories as follows:**


```
Calories = θ0 + θ1 × [ Carbohydrates ] + θ2 × [ Protein ] + θ3 × [ Total Fat ]
```
**Report the values of θ0,	θ1,	θ2 and θ3. Briefly describe your interpretation of these values, i.e., what do
θ0,	θ1,	θ2 and θ3 represent? Explain these in terms of the features and labels**

````python
# split data to feature and target variables, Use carbohydrates, protein and total fat to predict calories
X = mcdonals_df_linear_reg[["Carbohydrates", "Protein", "Total Fat"]]
y = mcdonals_df_linear_reg[["Calories"]]
````

````python

# Initialize the model
linear_model = LinearRegression()

# Train the model of complete data
linear_model.fit(X, y)

# Get the coefficients
theta_0 = linear_model.intercept_[0]  # This is the intercept
theta_1, theta_2, theta_3 = linear_model.coef_[0]  # These are the coefficients for the features

# Print the values
print(f"Intercept (theta0): {round(theta_0, 2)}")
print(f"Carbohydrates coefficient (theta1): {round(theta_1, 2)}")
print(f"Protein coefficient (theta2): {round(theta_2, 2)}")
print(f"Total Fat coefficient (theta3): {round(theta_3, 2)}")

````

## **(b): Split data into 90-10 ratio**





````python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Initialize and train Linear Regression Model
linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train, y_train)

# Make Predictions
y_train_pred = linear_reg_model.predict(X_train)
y_test_pred = linear_reg_model.predict(X_test)
````

````python
# Calculate Mean Squared Error (MSE)
mse_training = mean_squared_error(y_train, y_train_pred)
mse_testing = mean_squared_error(y_test, y_test_pred)

print("The mean squared error (MSE) on training data:", round(mse_training, 2))
print("The mean squared error (MSE) on testing data:", round(mse_testing, 2))
````

### **(b): Performance Calculation and Reasons**

````python
difference_mse = abs(mse_training - mse_testing)
print("The difference between MSE of training and testing data:", round(difference_mse, 2))
````

### **Reasons:**
- The difference between MSE of training and test data is 4.72 relatively small, that means model is slightly overfitting.
- The possible reason could be that 90% data was used for training, that could cause the model to learn noise and outliers too.
- The test data is very samll 10%, that's not able to represent the overall distribution of data.
- The model might have memorized the numbers rather than making generalized patterns.

# **Logistic Regression**

````python

# read diabetes data as pandas dataframe
diabetes_data_path = "/content/diabetes.csv"
diabetes_df = pd.read_csv(diabetes_data_path)

# display first 5 rows
diabetes_df.head()
````

````python
# Looking for missing data
diabetes_df.isnull().sum()
````

````python
# remove "Outcome" from features column
feature_columns = diabetes_df.columns.tolist().copy()
feature_columns.remove("Outcome")

# set target column as "Outcome"
target_column = "Outcome"
````

## **(a):  Train a set of logistic regression models with different values of `C (𝑐 = 0.01,0.1,1,10,100)` for both L1 and L2 penalties.**

````python
from sklearn.preprocessing import StandardScaler

# Split data into training and test sets with 70-30 ratio
X_train_diabetic, X_test_diabetic, y_train_diabetic, y_test_diabetic = train_test_split(diabetes_df[feature_columns], diabetes_df[target_column], test_size=0.3, random_state=42)

# Scale features for logistic regression
scaler = StandardScaler()
X_train_scaled_diabetic = scaler.fit_transform(X_train_diabetic)
X_test_scaled_diabetic = scaler.transform(X_test_diabetic)
````

````python
# Different values of C
C_values = [0.01, 0.1, 1, 10, 100]

# for smaller dataset and binary classification "liblinear" is efficient for L1 and "lbfgs" for L2 regularization.
penalty_solver = {"l1": "liblinear", "l2": "lbfgs"}

# Store coefficients and scores
l1_coefficients = list()
l2_coefficients = list()
l1_training_scores = list()
l2_training_scores = list()
l1_testing_scores = list()
l2_testing_scores = list()

# Train model with L1 and L2 penalties with their respectives solver
for C in C_values:
    for penalty, solver in penalty_solver.items():

        # Train logistic regression model
        logistics_model = LogisticRegression(C=C, penalty=penalty, solver='liblinear')
        logistics_model.fit(X_train_scaled_diabetic, y_train_diabetic)

        # Model Scores
        training_score = logistics_model.score(X_train_scaled_diabetic, y_train_diabetic)
        testing_score = logistics_model.score(X_test_scaled_diabetic, y_test_diabetic)

        # Model Accuracy
        y_pred_diabetic = logistics_model.predict(X_test_scaled_diabetic)
        accuracy = accuracy_score(y_test_diabetic, y_pred_diabetic)

        # Feature weights
        coefficients = logistics_model.coef_[0]

        if penalty == "l1":
            l1_coefficients.append(coefficients)  # Store L1 coefficients
            l1_training_scores.append(training_score)  # Store L1 training scores
            l1_testing_scores.append(testing_score)  # Store L1 testing scores
        else:
            l2_coefficients.append(coefficients)  # Store L2 coefficients
            l2_training_scores.append(training_score)  # Store L2 training scores
            l2_testing_scores.append(testing_score)  # Store L2 testing scores

````

## **(b): Affect of C on feature coefficients and model's performance:**

````python
# Convert to a numpy array for fetching of data based on index - 2D array
l1_coefficients_array = np.array(l1_coefficients)
l2_coefficients_array = np.array(l2_coefficients)

# Plotting
plt.figure(figsize=(15, 6))

# Plot L1 regularization
for i, feature in enumerate(feature_columns):
    # fetch each feature column coefficients from l1_coefficients_array 2D array
    plt.plot(C_values, l1_coefficients_array[:, i], label=f"L1 - {feature}", marker='o', color='red')

# Plot L2 regularization
for i, feature in enumerate(feature_columns):
    # fetch each feature column coefficients from l2_coefficients_array 2D array
    plt.plot(C_values, l2_coefficients_array[:, i], label=f"L2 - {feature}", marker='x', color='blue')

# Adding labels and title
plt.xscale('log')  # Log scale for C values
plt.xlabel('Regularization Strength (C = 1/λ)')
plt.ylabel('Feature Weights (Coefficients)')
plt.title('Feature Weights vs Regularization Strength (C = 1/λ) for L1 and L2 Regularization')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
````

### **Feature Weight vs Regularization Strength:**
- **L1 Regularization:**
  - When **C is very small**, this is **High Regularization**, model makes **most of the feature coefficient to 0**, performing **strong feature selection**.

  - When **C increases, Regularization weakens**, and **more feature coefficients become non-zero**, performing **moderate feature selection**.

  - When **C is very high**, **Negligible Regularization**, **almost all coefficients become non-zero**, performing **poor feature selection.**


- **L2 Regularization:**

  - When **C is very small**, **High Regularization**, **all feature are very close to 0, but none of them is exactly 0.** The model is simple.

  - When **C increases**, **Regularization weakens**, and **feature coefficients start growing**.

  - When **C is very high**, **Negligible Regularization**, and the **feature coefficients become large**, model becomes complex.

````python
# Plotting the scores
plt.figure(figsize=(15, 5))

# Plot L1 regularization scores
plt.plot(C_values, l1_training_scores, label='L1 Training Score', marker='o', color='b', markersize=6)
plt.plot(C_values, l1_testing_scores, label='L1 Testing Score', marker='x', color='r', markersize=6)

# Plot L2 regularization scores
plt.plot(C_values, l2_training_scores, label='L2 Training Score', marker='o', linestyle='--', color='b', markersize=6)
plt.plot(C_values, l2_testing_scores, label='L2 Testing Score', marker='x', linestyle='--', color='r', markersize=6)


# Adding labels and title
plt.xscale('log')  # Log scale for C values
plt.xlabel('Regularization Strength (C = 1/λ)')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Regularization Strength (C = 1/λ) for L1 and L2 Regularization')
plt.legend(loc='best')
plt.grid(True)
plt.show()
````

### **Accuracy vs Regularization Strength:**
- **L1 Regularization:**
  - When **C is very small**, there is **High Regularization**, model pushed most of the feature coefficients to 0, and **accuracy of both training and test data tends to be low**.
  - When **C increases**, **Moderate Regularization**, more feature coefficients become non-zero, so **accuracy of both training and test data increases**.
  - When **C is very high**, **Negligible Regularization**, model is less regularized and overfits the training data, **resulting in the saturation of accuracy on both training and test data**.

- **L2 Regularization:**
  - When **C is very small**, this is **High Regularization**, almost **all feature coeffients exists** and are **small close to 0**, gives **much better accuracy than L1 regularization on both training and testing data** at this range of C.
  - When **C increases**, **Moderate Regularization**, **feature coefficients start to grow**, and **accuracy on both the training and test starts to improves** as the model finds a good balance between simplicity and complexity.
  - When **C is very high**, **Negligible Regularization**, the model is near to no regularization, which leads to overfitting. **Accuracy on both training and testing starts to saturate.**
